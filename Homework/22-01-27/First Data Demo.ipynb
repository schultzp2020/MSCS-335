### Linear Regession
$acceleration = m_1(mpg) + m_2(cylinders) + m_3(displacement) + m_4(weight) + m_5(year)$

- $mpg$: MPG can inversely relate to acceleration
- $cylinders$: Cylinders can directly relate to acceleration
- $displacement$: Displacement can inversely relate to acceleration
- $weight$: Weight can inversely relate to acceleration
- $year$: Year can directly relate to acceleration

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(df[['mpg', 'cylinders', 'displacement', 'weight', 'year']], df['acceleration'])
model.coef_
```

Below uses the average of each of the independent variables to predict the acceleration.

```python
x = np.array([[23.515869, 5.458438, 193.532746, 2970.261965, 75.994962]])

model.predict(x)
```
