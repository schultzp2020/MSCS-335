The dataset I chose was the NASA Kepler Exoplanet dataset. The data came from the Kepler space telescope which was launched by NASA in 2009 to discover Earth-sized exoplanets. To analyze the data, the telescope would transmit the data back to Earth, then the data would be analyzed to detect periodic dimming caused by exoplanets crossing in front of their star. The dataset was gathered from when the telescope launched to when the telescope was decommissioned in 2018. In total, Kepler observed 530,506 stars and 2,662 planets.

In the Kepler Exoplanet dataset, there are seven possible categories a column can fit. Identification Columns, Exoplanet Archive Information, Project Disposition Columns, Transit Properties, Threshold-Crossing Event (TCE) Information, Stellar Parameters, or KIC Parameters. The Identification Columns, Exoplanet Archive Information, and Project Disposition Columns categories are basic columns like the Kepler ID, Kepler Object of Interest (KOI) name / Kepler name, KOI disposition, KOI p-disposition, KOI score, and flag values. The KOI disposition exoplanet values can be CANDIDATE, FALSE POSITIVE, NOT DISPOSITIONED or CONFIRMED. The difference between disposition and p-disposition is that disposition is according to scholarly sources and p-disposition is according to statistical analysis. KOI score is an amalgamation of predictors with a value between 0 and 1 that indicates confidence in the KOI disposition with a higher value being more confident. Lastly the flag values are either TRUE or FALSE depending on the circumstances of the observed potential exoplanet. E.g., the Not Transit-Like Flag is a KOI whose light curve is not consistent with that of a transiting planet. 

The Transit properties category are columns regarding how a potential exoplanet revolves around its star. KOI period is the interval between consecutive planetary transits. KOI time0bk is the time corresponding to the center of the first detected transit starting zero time of January 1st, 2009, UTC. KOI impact is the sky-projected distance between the center of the stellar disc and the center of planet disc normalized. KOI duration is the duration of the observed transits. KOI depth is the fraction of stellar flux (light) lost at the minimum of the transit. KOI planet radius is the radius of the planet. KOI equilibrium temperature is the approximation of the temperature of the planet. KOI insolation flux is another way to give equilibrium temperature. All these values also have a positive and negative error corresponding to err1 and err2 respectively.

The Threshold-Crossing Event (TCE) Information category are columns after analysis regarding the potential exoplanetâ€™s transit. KOI model snr is the transit depth normalized. KOI TCE planet number is the TCE planet number federated to the KOI. KOI TCE delivery name is the delivery name regarding the planet number. 

The Stellar Parameters category are columns regarding the star parameters. KOI stellar effective temperature is the photospheric temperature of the star. KOI stellar surface gravity is the base-10 logarithm of the acceleration due to gravity at the surface of the star. KOI stellar radius is the photospheric radius of the star. Lastly KIC Parameters category are columns regarding the Kepler space telescope with KIC right Ascension, KIC declination, and Kepler band. There is not much information on this last category.
