### Linear Regession
$sales = m_1(tv) + m_2(radio) + m_3(paper)$

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(df[['TV', 'radio', 'newspaper']], df['sales'])
model.coef_
```

```python
x = np.array([[100, 50, 75]])

model.predict(x)
```
