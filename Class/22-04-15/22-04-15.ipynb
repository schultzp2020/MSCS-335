```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('forSVD.csv', index_col=0)
```

```python
df.describe()
```

```python
pd.plotting.scatter_matrix(df)
```

```python
pca = PCA()
pca.fit(df)
```

```python
for s in pca.singular_values_:
  print(s/pca.singular_values_[0])
```

```python
for ev in pca.explained_variance_:
  print(ev*100)
```

```python
pca = PCA(n_components=3)
pca.fit(df)
```

```python
for s in pca.singular_values_:
  print(s/pca.singular_values_[0])
```

```python
for ev in pca.explained_variance_:
  print(ev*100)
```

```python
g = pd.DataFrame(pca.transform(df), columns=[f'pc{i}' for i in range(pca.n_components)])
g.shape
```

```python
g
```

```python
pd.plotting.scatter_matrix(g)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(g['pc0'], g['pc1'], g['pc2'])
ax.view_init(270, 0)
```
