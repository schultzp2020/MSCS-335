```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x1 = np.random.rand(2000, 2) * np.array([5, 2]) + np.array([-10, 5])
x2 = (np.random.randn(3000, 2) * np.array([8, 1])) @ np.array([[0.8, -0.6], [0.6, 0.8]]) + np.array([20, 10])
t3 = np.linspace(0, 8*np.pi, 1000)
x3 = np.array([t3 * np.cos(t3), t3 * np.sin(t3)]).T / 5 + np.array([-5, -5])
X = np.vstack([x1, x2, x3])

N = X.shape[0]
np.random.shuffle(X)

plt.scatter(X[:,0], X[:,1], marker='.')
print(X.shape)
```

```python
c = KMeans(4, random_state=2022)
labels = c.fit_predict(X)
```

```python
fig, ax = plt.subplots(1, figsize=(4, 4), dpi=100)
for i in range(c.n_clusters):
    ax.scatter(X[labels==i,0], X[labels==i,1], marker='x')
ax.scatter(c.cluster_centers_[:,0], c.cluster_centers_[:,1], color='black', marker='o')
ax.set_aspect(1.0)
```

```python
D = c.transform(X)
M = c.transform(c.cluster_centers_)
```

```python
fig, ax = plt.subplots(1, figsize=(4, 4), dpi=100)
for i in range(c.n_clusters):
    ax.scatter(D[labels==i,0], D[labels==i,1], marker='x')
ax.scatter(M[:,0], M[:,1], color='black', marker='o')
ax.set_aspect(1.0)
```
