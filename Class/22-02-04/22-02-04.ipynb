```python
import pandas as pd
import numpy as np
import numpy.linalg as la
from itertools import combinations_with_replacement as cwr
```

```python
df = pd.read_csv('https://statlearning.com/s/Advertising.csv', index_col=0)
```

```python
df
```

```python
Y = df['sales']
```

```python
X = df.drop(columns=['sales'])
```

```python
M = la.solve(X.T@X, X.T@Y)
```

```python
M
```

```python
X_copy = X.copy()
for term in cwr(X.columns, 4):
  name = '*'.join(term)
  column = 1
  columns = [X[colname] for colname in term]
  for col in columns:
    column *=col
  X_copy[name] = column
```

```python
X_copy
```

```python
X_copy = X.copy()
for term in cwr(X.columns, 4):
  name = '*'.join(term)
  column = X_copy[list(term)].apply(np.prod, axis=1)
  X_copy[name] = column
```
