```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
x = np.linspace(0, 4*np.pi, 200)
y = np.sin(x) + np.random.randn(200)*0.2
df = pd.DataFrame({'x1': x, 'y': y})
```

```python
X = df.drop(columns=['y'])
Y = df['y']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
```

```python
fig, ax = plt.subplots(1, figsize=(4,3), dpi=100)
ax.scatter(X_train, Y_train, color='blue', marker='x')
ax.scatter(X_test, Y_test, color='orange', marker='+')
plt.show()
```

```python
model = DecisionTreeRegressor(max_depth=3)
model.fit(X_train, Y_train)
print(f"training score is {model.score(X_train, Y_train)}")
model.fit(X_train, Y_train)
print(f"training score is {model.score(X_test, Y_test)}")
```

```python
fig, ax = plt.subplots(1, figsize=(4,3), dpi=100)
# ax.scatter(X_train, Y_train, color='blue', marker='x')
# ax.scatter(X_train, model.predict(X_train), color='green', marker='.')
ax.scatter(X_test, model.predict(X_test), color='red', marker='.')
ax.scatter(X_test, Y_test, color='orange', marker='+')
plt.show()
```
